1. Prepare the sample csv file for car model and their specification, the column names would be   
  a. Company Name  
  b. Model Name  
  c. Fuel Type (For e.g. Petrol, Gas, Diesel, EV)  
  d. Body Style (for e.g. Hatchback, Sedan, SUV)  
  e. Car Length  

```python
# Create the Dataframe to convert the data into csv
data = {"Company_Name":["Ford","Hyundai","Kia"],"Model_Name":["EcoSport","i20","Seltos"],"Fuel_Type":["Diesel","Petrol","Diesel"],"Body_Style":["SUV","Hatchback","SUV"],"Car_Length":["3998mm","3995mm","4315mm"]}
df = pd.DataFrame(data)
df
```

```python
# create csv file
df.to_csv("car.csv")
```

2. Prepare the sample excel file for the car prizing and loan amount, the column names would be  
a. Company Name     
b. Model Name  
c. On road pricing  
d. Loan amount  
e. Monthly EMI  
f. Interest Rate  
g. Monthly Principal  
h. Monthly Interest  

```python
# Create the Dataframe to convert the data into csv
data = {"Company_Name":["Ford","Hyundai","Kia"],"Model_Name":["EcoSport","i20","Seltos"],"On_Road_Pricing":["800000","1200000","1100000"],"Loan_Amount":["500000","1000000","1000000"]," Monthly_EMI":["52000","45000","35000"],"Interest_Rate":["8","6","9"],"Monthly_Principal":["8000","9000","7800"]," Monthly_Interest":["10000","9000","8900"]}
dataset = pd.DataFrame(data)
dataset
```

```python
# Create excel file
dataset.to_excel("car_price.xlsx")
```

3. Write the code to read the csv file and the excel file and convert them into the Data frame.

```python
# code to read the csv file
csv_data = pd.read_csv("car.csv")
csv_data
```

```python
# code to read the excel file
excel_data = pd.read_excel("car_price.xlsx")
excel_data
```

4. Merge the two Data frame on the basis of primary key

```python
merged_data = csv_data.merge(excel_data, on="Company_Name")
merged_data
```
